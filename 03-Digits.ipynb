### Partição dos dados

Realiza-se um split randomico dos dados, com 60% dos dados para treino e 40% para teste.

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(digitos.data,
                                                                     digitos.target,
                                                                     test_size=.4)
```

### Treinamento do modelo

#### O Scikit-learn fornece um conjunto de modelos, com uma interface com os seguintes métodos:

- **Presentes em todos estimators:**
    - **fit():** é o método para o qual é passado o dataset. No caso dos supervised estimatores, possui dois parâmetros, o primeiro sendo os features, o segundo sendo os labels/target.
- **Presentes nos supervised estimators:**
    - **predict():** após treinado, retorna os labels para o conjunto de features passado.
    - **predict_proba():** retorna a probabilidade para cada label, em modelos probabilísticos.
    - **score():** retorna um valor de 0 a 1 representando quão boa é a predição.
- **Presentes nos unsupervised estimators:**
    - **transform():** lembrar do PCA. Apenas transforma, usando os parâmetros do modelo.
    - **fit_transform():** lembrar do PCA. Ajusta os parãmetros do modelo aos dados e os transforma.

```python
model = svm.SVC(gamma=0.001, C=100.0)
model.fit(X_train, y_train)

pickle.dump(model, open('model', 'w'))
```

### Testando o modelo

```python
y_predito = model.predict(X_test)
```

#### Métricas de classificação

```python
print metrics.classification_report(y_test, y_predito)
```

#### Matriz de confusão

Matriz cujas colunas representam as classes preditas, e as linhas as classes reais

```python
cm = metrics.confusion_matrix(y_test, y_predito)

cms = np.zeros((11, 11))
cms[0, :] = np.arange(-1, 10, 1)
cms[:, 0] = np.arange(-1, 10, 1).reshape((11,))
cms[1:, 1:] = cm
print cms
```
